```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu
models = ['Logistic Regression', 'SGDClassifier', 'CNN']
f1_sample = [0.26700788656480173, 0.1706208501822783, 0.3967057347152688]
f1_micro = [0.3340496730420901, 0.21409992921893523, 0.4482074160893031]
f1_macro = [0.1910027212507754, 0.08448656141383039, 0.36230866038772125]

x = np.arange(len(models))
width = 0.3

# Tạo biểu đồ
bars1 = plt.bar(x - width, f1_sample, width, label='F1 Sample', color='skyblue')
bars2 = plt.bar(x, f1_micro, width, label='F1 Micro', color='lightgreen')
bars3 = plt.bar(x + width, f1_macro, width, label='F1 Macro', color='salmon')

for bar in bars1:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.3f}', 
             ha='center', va='bottom', fontsize=10)

for bar in bars2:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.3f}', 
             ha='center', va='bottom', fontsize=10)

for bar in bars3:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.3f}', 
             ha='center', va='bottom', fontsize=10)

plt.xlabel('Giải pháp')
plt.ylabel('F1 Scores')
plt.title('So sánh độ đo F1 giữa các giải pháp')
plt.xticks(x, models)
plt.ylim(0, 1) 
plt.legend()

plt.tight_layout()
plt.show()
```

## So sánh các tham số khi sử dụng CNN

```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu
models = ['CNN(99k bản ghi, 6704 tags)', 'CNN(73383 bản ghi, 10214 tags)']
f1_sample = [0.3967057347152688,0.3685979702488526]
f1_micro = [0.4482074160893031,0.4232423545684525]
f1_macro = [0.36230866038772125,0.2998352168245054]

x = np.arange(len(models))
width = 0.2

# Tạo biểu đồ
bars1 = plt.bar(x - width, f1_sample, width, label='F1 Sample', color='skyblue')
bars2 = plt.bar(x, f1_micro, width, label='F1 Micro', color='lightgreen')
bars3 = plt.bar(x + width, f1_macro, width, label='F1 Macro', color='salmon')

for bar in bars1:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.3f}', 
             ha='center', va='bottom', fontsize=10)

for bar in bars2:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.3f}', 
             ha='center', va='bottom', fontsize=10)

for bar in bars3:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.3f}', 
             ha='center', va='bottom', fontsize=10)

plt.xlabel('Giải pháp')
plt.ylabel('F1 Scores')
plt.title('So sánh độ đo F1 giữa các lần thực nghiệm với CNN')
plt.xticks(x, models)
plt.ylim(0, 1) 
plt.legend()

plt.tight_layout()
plt.show()
```
